```python
import datetime
import calendar
import random

# Empleados y sus roles, sin Anthony Cerdas
empleados = [
    ('Elio Cortés', 'Desarrollo'), ('Keilly Chinchilla', 'Desarrollo'),
    ('Yulian Ulloa', 'Desarrollo'), ('Andrés Oreamuno', 'Desarrollo'),
    ('Alejandra González', 'Administrativo'), ('Yendri Leandro', 'Administrativo'),
    ('Otto Rivera', 'Administrativo'), ('Juan Carlos Bertsch', 'Administrativo'),
    ('Thomas Bertsch', 'Administrativo'), ('Maikel Granados', 'Soporte'),
    ('Róger Moreira', 'Soporte'), ('Rodney Mathieu', 'Soporte'),
]

# Fechas festivas de Costa Rica en 2024
festivos_2024 = {
    datetime.date(2024, 1, 1), datetime.date(2024, 3, 28), datetime.date(2024, 3, 29),
    datetime.date(2024, 4, 15), datetime.date(2024, 5, 1), datetime.date(2024, 7, 29),
    datetime.date(2024, 8, 2), datetime.date(2024, 8, 15), datetime.date(2024, 9, 15),
    datetime.date(2024, 12, 25)
}

def es_dia_laborable(fecha):
    return fecha.weekday() < 5 and fecha not in festivos_2024

def obtener_proxima_semana():
    hoy = datetime.date.today()
    inicio_semana = hoy + datetime.timedelta(days=-hoy.weekday(), weeks=1)
    return [inicio_semana + datetime.timedelta(days=i) for i in range(5) if es_dia_laborable(inicio_semana + datetime.timedelta(days=i))]

def distribucion_inteligente(proxima_semana):
    asignaciones = {empleado: [] for empleado, _ in empleados}
    dias_laborables = obtener_proxima_semana()

    # Asegurar distribución equitativa
    roles = {'Desarrollo': 2, 'Administrativo': 1, 'Soporte': 1}

    for dia in dias_laborables:
        empleados_dia = []
        while not empleados_dia:
            for empleado, rol in empleados:
                if len(asignaciones[empleado]) < roles[rol]:
                    if random.random() < 0.5:  # Probabilidad de asignación
                        asignaciones[empleado].append(dia)
                        empleados_dia.append(empleado)
        
        # Asegurar al menos un empleado por día
        if not empleados_dia:
            empleado_aleatorio = random.choice(list(asignaciones.keys()))
            asignaciones[empleado_aleatorio].append(dia)

    # Ordenar las fechas asignadas
    for empleado in asignaciones:
        asignaciones[empleado] = sorted(set(asignaciones[empleado]))

    return asignaciones

def imprimir_lista(asignaciones):
    for empleado, dias in sorted(asignaciones.items()):
        print(f"{empleado}:")
        for dia in dias:
            print(f"  - {dia.strftime('%Y-%m-%d')}")

def imprimir_tabla_para_excel(asignaciones):
    print("Empleado,Fecha 1,Fecha 2")
    for empleado, dias in asignaciones.items():
        fecha1 = dias[0].strftime('%Y-%m-%d') if len(dias) > 0 else "N/A"
        fecha2 = dias[1].strftime('%Y-%m-%d') if len(dias) > 1 else "N/A"
        print(f"{empleado},{fecha1},{fecha2}")

def imprimir_tabla_completa(asignaciones):
    print("Empleado,", end="")
    dias_semana = obtener_proxima_semana()
    for dia in dias_semana:
        print(f"{dia.strftime('%A %d/%m')},", end="")
    print()

    for empleado, _ in empleados:
        print(f"{empleado},", end="")
        for dia in dias_semana:
            if dia in asignaciones[empleado]:
                print("X,", end="")
            else:
                print("-,", end="")
        print()

def main():
    proxima_semana = obtener_proxima_semana()
    asignaciones = distribucion_inteligente(proxima_semana)
    print("\nAsignaciones de días para empleados:")
    imprimir_lista(asignaciones)
    print("\nTabla de asignaciones para la próxima semana (para Excel):")
    imprimir_tabla_para_excel(asignaciones)
    print("\nTabla completa de asignaciones para la próxima semana:")
    imprimir_tabla_completa(asignaciones)

if __name__ == "__main__":
    main()
```
